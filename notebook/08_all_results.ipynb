```python
import pandas as pd

# Load all 4 result files
heart = pd.read_csv('../results/heart_results.csv', index_col=0)
credit = pd.read_csv('../results/credit_results.csv', index_col=0)
german = pd.read_csv('../results/german_credit_results.csv', index_col=0)
insurance = pd.read_csv('../results/insurance_results.csv', index_col=0)

# Add dataset column
heart['Dataset'] = 'Heart Disease'
credit['Dataset'] = 'Credit Card'
german['Dataset'] = 'German Credit'
insurance['Dataset'] = 'Insurance'

# Combine
all_results = pd.concat([heart, credit, german, insurance])
all_results = all_results.reset_index().rename(columns={'index': 'Model'})

print("=== ALL EXPERIMENTAL RESULTS ===")
print(all_results)

# Save master table
all_results.to_csv('../results/ALL_RESULTS.csv', index=False)

print("\n✅ ALL 4 DATASETS COMPLETE!")
print("\nQuick summary:")
for dataset in ['Heart Disease', 'Credit Card', 'German Credit', 'Insurance']:
    subset = all_results[all_results['Dataset'] == dataset]
    best = subset.loc[subset['Accuracy'].idxmax()]
    print(f"{dataset}: Best = {best['Model']} ({best['Accuracy']:.4f})")
```
