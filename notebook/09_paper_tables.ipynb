```python
import pandas as pd

df = pd.read_csv('../results/ALL_RESULTS.csv')

print(df.head())
print("\nShape:", df.shape)
```

```python
table1 = df.pivot_table(
    values=['Accuracy', 'F1 Score', 'ROC-AUC'],
    index='Model',
    columns='Dataset',
    aggfunc='mean'
)

print("TABLE 1: Model Performance Across Datasets")
print(table1.round(4))

table1.to_csv('../results/Table1_Performance.csv')
```

```python
datasets_info = {
    'Dataset': ['Heart Disease', 'Credit Card', 'German Credit', 'Insurance'],
    'Samples': [303, 30000, 999, 1338],
    'Features': [13, 23, 24, 6],
    'Best Model': []
}

for dataset in datasets_info['Dataset']:
    subset = df[df['Dataset'] == dataset]
    best_model = subset.loc[subset['Accuracy'].idxmax(), 'Model']
    datasets_info['Best Model'].append(best_model)

table2 = pd.DataFrame(datasets_info)

print("\nTABLE 2: Dataset Characteristics")
print(table2)

table2.to_csv('../results/Table2_Datasets.csv', index=False)
```
