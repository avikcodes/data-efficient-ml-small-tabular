```python
import pandas as pd
import numpy as np

# Load correctly (no header)
df = pd.read_csv('../datasets/heart.csv', header=None)

# Strategy 1: If < 5% missing, drop those rows
if df.isnull().sum().sum() / len(df) < 0.05:
    df_clean = df.dropna()
    print(f"Dropped {len(df) - len(df_clean)} rows with missing values")
else:
    # Strategy 2: Fill numeric columns with median
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        df[col].fillna(df[col].median(), inplace=True)
    
    # Fill categorical with mode
    cat_cols = df.select_dtypes(include=['object']).columns
    for col in cat_cols:
        df[col].fillna(df[col].mode()[0], inplace=True)
    
    df_clean = df
    print("Filled missing values")

print(f"Clean dataset shape: {df_clean.shape}")
```

```python
target_col = 13

X = df_clean.drop(columns=[target_col])
y = df_clean[target_col]

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"\nFeature columns: {X.columns.tolist()}")
```

```python
# Check if there are any categorical (non-numeric) columns
print("Data types:")
print(X.dtypes)

# If you have categorical columns, encode them
cat_columns = X.select_dtypes(include=['object']).columns

if len(cat_columns) > 0:
    print(f"\nCategorical columns found: {cat_columns.tolist()}")
    from sklearn.preprocessing import LabelEncoder
    
    le = LabelEncoder()
    for col in cat_columns:
        X[col] = le.fit_transform(X[col])
    print("Encoded categorical variables")
else:
    print("No categorical columns - all numeric!")
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize all features (mean=0, std=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("Features normalized")
print(X_scaled.describe())
```

```python
# Save cleaned data
X_scaled.to_csv('../datasets/heart_X_clean.csv', index=False)
y.to_csv('../datasets/heart_y_clean.csv', index=False)

print("✅ Cleaned data saved!")
print(f"   - Features: {X_scaled.shape}")
print(f"   - Target: {y.shape}")
print(f"   - Ready for modeling tomorrow")
```
