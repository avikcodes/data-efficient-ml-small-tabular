```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load dataset
df = pd.read_csv('../datasets/credit_card.csv', header=1)


print(df.head())
print(f"Shape: {df.shape}")
print(f"Missing values:\n{df.isnull().sum()}")

# ---- FIX TARGET COLUMN NAME ----
target_col = 'default payment next month'

print(f"\nTarget distribution:\n{df[target_col].value_counts()}")

# Drop ID column (useless for ML)
if 'ID' in df.columns:
    df = df.drop(columns=['ID'])

# Handle missing values
if df.isnull().sum().sum() > 0:
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        df[col].fillna(df[col].median(), inplace=True)

# Separate X and y
X = df.drop(columns=[target_col])
y = df[target_col]

# Encode categorical columns
cat_cols = X.select_dtypes(include=['object']).columns
for col in cat_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])

# Scale features
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Save cleaned dataset
X_scaled.to_csv('../datasets/credit_X_clean.csv', index=False)
y.to_csv('../datasets/credit_y_clean.csv', index=False)

print("\nDataset 2 cleaned and saved!")
print(f"Features: {X_scaled.shape}")
```
